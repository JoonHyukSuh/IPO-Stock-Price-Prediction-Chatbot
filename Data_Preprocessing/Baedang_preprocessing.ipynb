```python
import pandas as pd
df = pd.read_csv('배당정보.csv',encoding= 'euc-kr')
df.rename(columns = df.iloc[0],inplace=True)
df.drop(df.index[0],inplace = True)

df
```

```python
df.rename(columns = {'업종별':'업종별배당율',
                    '현금배당':'현금배당액면가'},inplace = True)
df = df.iloc[1:,:8]
df.loc[df['업종']=='-'] = '기타'
```

```python
temp = pd.DataFrame(df.groupby(['업종','회사명','업종별배당율','현금배당액면가'])['사업연도'].max())
temp = temp.reset_index(level=['업종별배당율'])
temp.drop(['사업연도'],axis = 1,inplace = True)
temp = temp.reset_index(level=['현금배당액면가'])
temp
```
